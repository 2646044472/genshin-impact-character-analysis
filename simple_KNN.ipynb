```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_X = [
    [1.3, 6],
    [3.5, 5],
    [4.2, 2],
    [5.3, 3],
    [2, 9],
    [5,7.5],
    [7.2,4],
    [8.1,8],
    [9,2.5]
]

#特征值
data_feature = [0, 0, 0, 0, 1, 1, 1, 1, 1]
```

```python
data_X = np.array(data_X)
```

```python
data_feature = np.array(data_feature)
```

```python
data_new = np.array([4,5])
```

```python
plt.scatter(data_X[data_feature==0][:,0],data_X[data_feature==0][:,1], color="red", marker="x")
plt.scatter(data_X[data_feature==1][:,0],data_X[data_feature==1][:,1], color="red", marker="o")
plt.scatter(data_new[0],data_new[1], color="red", marker="^")
```

```python
distance = [np.sqrt(np.sum((data-data_new)**2)) for data in data_X]
distance
```

```python
sort_index = np.argsort(distance)
sort_index
```

```python
k = 5
```

```python
first_k = [data_feature[i] for i in sort_index[:k]]
first_k = pd.DataFrame(first_k)
s = first_k.value_counts()
s
```

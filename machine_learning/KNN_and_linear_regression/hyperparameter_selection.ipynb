```python
iris = datasets.load_iris()
```

```python
y = iris.target
y.shape
```

# 拆分数据集

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,random_state=310,stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

穷举搜索

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(
    n_neighbors=3,
    weights='distance', #'uniform'
    p = 2
)
```

```python
best_score = -1
best_n = -1
best_weight = ''
best_p = -1
for n in range(1,20):
    for weight in ['uniform','distance']:
        for p in range(1,7):
            neigh = KNeighborsClassifier(
                n_neighbors = n,
                weights = weight, #'uniform'
                p = p
            )
            neigh.fit(X_train,y_train)
            score = neigh.score(X_test,y_test)
            if score > best_score:
                best_score = score
                best_n = n
                best_weight = weight
                best_p = p
                
best_score, best_n, best_weight, best_p
```

sklearn自带方法

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors':[n for n in range(1,20)],
    'weights':['uniform','distance'],
    'p':[p for p in range(1,7)]
}
```

```python
grid = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=params,
    n_jobs=-1
)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
grid.best_estimator_.score(X_test,y_test)
```

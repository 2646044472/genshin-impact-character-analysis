```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-4,2,size=(100))    #生成-4到2的100个随机数
y = 2*x**2 + 4*x + 3 + np.random.randn(100)
X = x.reshape (-1,1)
```

```python
plt.scatter(x,y)
```

```python
X[:5]
```

```python
X_new = np.hstack([X,X**2])
```

```python
X_new[:5]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_new,y)
```

```python
y_predict = lin_reg.predict(X_new)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color="red")
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color="red")
```

```python
lin_reg.coef_
```

sklearn中的PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_features = PolynomialFeatures(degree=2)
```

```python
X_poly = polynomial_features.fit_transform(X)    # 生成一个特征矩阵,包含所有有可能的多项式组合，不超过degree
```

```python
X_poly[:5]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_poly,y)
```

```python
y_predict = lin_reg.predict(X_poly)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color="red")
```
